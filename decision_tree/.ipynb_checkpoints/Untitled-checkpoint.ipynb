```python
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np

vars_to_use = ['econ_gradient', 'I_perc',
       'R_perc', 'H_perc', 'ICU_utilization', 'xi', 'infections_received',
        'deaths_given', 'death_prob']
tree_depth = 4
```

```python
data = pd.read_csv("dataset.csv")
data['infections_received'] = data['infections_received']*1000
data['infections_given'] = data['infections_given']*1000
print(data.columns)
print(len(data))
```

```python
clf = tree.DecisionTreeRegressor(max_depth=tree_depth)
clf = clf.fit(data[vars_to_use], data["lockdown"])
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=1200)
tree.plot_tree(clf, feature_names=vars_to_use) 
fig.savefig('tree.png')
```
